```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

print("Libraries ready hain! ")
```

```python
data = {
    'Vendor_Name': ['Vendor_A', 'Vendor_B', 'Vendor_C', 'Vendor_D', 'Vendor_E'],
    'Items':       ['Toys', 'Food', 'Clothes', 'Shoes', 'Furniture'],
    'Price':       [500, 300, 700, 1200, 900],
    'Delivery_Days': [3, 1, 5, 7, 4],
    'Quality_Rating': [8, 9, 7, 6, 8]
}

df = pd.DataFrame(data)
df.to_csv('vendor_data.csv', index=False)
print("CSV ban gayi! ")
print(df)
```

```python
df = pd.read_csv('vendor_data.csv')

print("=== Price Analysis ===")
sasta = df.loc[df['Price'].idxmin(), 'Vendor_Name']
mahanga = df.loc[df['Price'].idxmax(), 'Vendor_Name']
print(f"Sabse Sasta  : {sasta} ({df['Price'].min()} Rs)")
print(f"Sabse Mahanga: {mahanga} ({df['Price'].max()} Rs)")
print()

print("=== Delivery Analysis ===")
fast = df.loc[df['Delivery_Days'].idxmin(), 'Vendor_Name']
slow = df.loc[df['Delivery_Days'].idxmax(), 'Vendor_Name']
print(f"Sabse Fast: {fast} ({df['Delivery_Days'].min()} Days)")
print(f"Sabse Slow: {slow} ({df['Delivery_Days'].max()} Days)")
print()

print("=== Quality Analysis ===")
best = df.loc[df['Quality_Rating'].idxmax(), 'Vendor_Name']
worst = df.loc[df['Quality_Rating'].idxmin(), 'Vendor_Name']
print(f"Best Quality : {best} ({df['Quality_Rating'].max()})")
print(f"Worst Quality: {worst} ({df['Quality_Rating'].min()})")
```

```python
import matplotlib.pyplot as plt
df = pd.read_csv('vendor_data.csv')
plt.figure(figsize=(8, 5))
plt.bar(df['Vendor_Name'], df['Price'],
        color=['red', 'green', 'orange', 'blue', 'purple'])

plt.title('Vendor Price Comparison', fontsize=14)
plt.xlabel('Vendor Name', fontsize=12)
plt.ylabel('Price in Rs', fontsize=12)

for i, val in enumerate(df['Price']):
    plt.text(i, val + 10, str(val) + ' Rs',
             ha='center', fontsize=11)

plt.tight_layout()
plt.show()

print("Sabse Sasta  :", df.loc[df['Price'].idxmin(), 'Vendor_Name'])
print("Sabse Mahanga:", df.loc[df['Price'].idxmax(), 'Vendor_Name'])
```

```python
import matplotlib.pyplot as plt
df = pd.read_csv('vendor_data.csv')

plt.figure(figsize=(8, 5))
plt.bar(df['Vendor_Name'], df['Delivery_Days'],
        color=['blue', 'green', 'orange', 'red', 'purple'])

plt.title('Vendor Delivery Days Comparison', fontsize=14)
plt.xlabel('Vendor Name', fontsize=12)
plt.ylabel('Delivery Days', fontsize=12)

for i, val in enumerate(df['Delivery_Days']):
    plt.text(i, val + 0.1, str(val) + ' Days',
             ha='center', fontsize=11)

plt.axhline(y=df['Delivery_Days'].mean(),
            color='black', linestyle='--', label='Average')
plt.legend()
plt.tight_layout()
plt.show()

print("Sabse Fast:", df.loc[df['Delivery_Days'].idxmin(), 'Vendor_Name'], "✅")
print("Sabse Slow:", df.loc[df['Delivery_Days'].idxmax(), 'Vendor_Name'], "❌")
```

```python
df = pd.read_csv('vendor_data.csv')

plt.figure(figsize=(8, 6))
colors = ['red', 'green', 'orange', 'blue', 'purple']
explode = [0, 0.1, 0, 0, 0]

plt.pie(df['Quality_Rating'],
        labels=df['Vendor_Name'],
        colors=colors,
        explode=explode,
        autopct='%1.1f%%',
        startangle=90,
        shadow=True)

plt.title('Vendor Quality Rating', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv('vendor_data.csv')
df['Score'] = df['Quality_Rating'] - (df['Delivery_Days'] * 0.5)

print("=" * 45)
print("   VENDOR PERFORMANCE ANALYSIS REPORT   ")
print("=" * 45)

print("\n PRICE SUMMARY")
print("-" * 45)
print(f"Sabse Sasta  : {df.loc[df['Price'].idxmin(), 'Vendor_Name']} ({df['Price'].min()} Rs)")
print(f"Sabse Mahanga: {df.loc[df['Price'].idxmax(), 'Vendor_Name']} ({df['Price'].max()} Rs)")
print(f"Average Price: {df['Price'].mean()} Rs")

print("\n DELIVERY SUMMARY")
print("-" * 45)
print(f"Sabse Fast: {df.loc[df['Delivery_Days'].idxmin(), 'Vendor_Name']} ({df['Delivery_Days'].min()} Days)")
print(f"Sabse Slow: {df.loc[df['Delivery_Days'].idxmax(), 'Vendor_Name']} ({df['Delivery_Days'].max()} Days)")

print("\n QUALITY SUMMARY")
print("-" * 45)
print(f"Best Quality : {df.loc[df['Quality_Rating'].idxmax(), 'Vendor_Name']} ({df['Quality_Rating'].max()})")
print(f"Worst Quality: {df.loc[df['Quality_Rating'].idxmin(), 'Vendor_Name']} ({df['Quality_Rating'].min()})")

print("\n OVERALL RANKING")
print("-" * 45)
ranking = df[['Vendor_Name', 'Score']].sort_values('Score', ascending=False)
ranking.index = range(1, len(ranking) + 1)
print(ranking)

winner = df.loc[df['Score'].idxmax(), 'Vendor_Name']
print("\n" + "=" * 45)
print(f" BEST VENDOR : {winner}")
print("=" * 45)
```
